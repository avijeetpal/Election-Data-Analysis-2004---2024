```python
# check Unique value and count them 
print(data1['State'].unique())
print(data1['State'].nunique())
```

```python
data1.loc[data1['Year'] == 2004, 'State'] = data1.loc[data1['Year'] == 2004, 'State']\
    .str.replace(r'CODE\s*:.*', '', regex=True)\
    .str.replace(r'S\.\s*', '', regex=True)\
    .str.replace(r'\d+', '', regex=True)\
    .str.strip()
print(data1.loc[data1['Year'] == 2009, 'State'].unique())
print(data1.loc[data1['Year'] == 2014, 'State'].unique())
print(data1.loc[data1['Year'] == 2019, 'State'].unique())
print(data1.loc[data1['Year'] == 2024, 'State'].unique())
print(data1['State'].unique())
print(data1['State'].nunique())
```

```python
# Change datatype of each column

data3['Year'] = data3['Year'].astype('int')
data3['Seats'] = data3['Seats'].astype('int')
data3['Total_Electors'] = data3['Total_Electors'].astype('int')
data3['Candidate_expen_per_Seat'] = pd.to_numeric(data3['Candidate_expen_per_Seat'].replace('[₹,?]', '', regex=True).astype(float))
data3['Max_Legal_Candidate_Expense'] = pd.to_numeric(data3['Max_Legal_Candidate_Expense'].replace('[₹,?]', '', regex=True).astype(float).astype(int))
data3['Legal_Cost_per_Voter'] = pd.to_numeric(data3['Legal_Cost_per_Voter'].replace('[₹,?]', '', regex=True).astype(float))
data3['Gove_Expense'] = pd.to_numeric(data3['Gove_Expense'].replace('[₹,?]', '', regex=True).astype(float).astype(int))
data3['Legal_Cost_per_Voter.1'] = pd.to_numeric(data3['Legal_Cost_per_Voter.1'].replace('[₹,?]', '', regex=True).astype(float))
print(data3.dtypes)
print(data3.head())
```
